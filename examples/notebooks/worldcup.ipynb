```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
plt.xkcd()
import math

first_year = 1990


years = [ y for y in list(range(first_year,2023,4)) if y not in [1942,1946]]

def awareness(t):
    adt = np.min([ abs(t-y) for y in years])
    return max(0.01,(t-first_year)/(2023-first_year)*0.2*math.exp(-30*adt))

ts = range(first_year,2023,1)
ys = [ awareness(t) for t in ts ]
plt.plot(ts,ys)
ax = plt.gca()
ax.set_ylim([0, 1])
plt.grid()
plt.title("Humanity's awareness of Monte Carlo")
```
